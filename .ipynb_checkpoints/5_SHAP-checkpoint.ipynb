Neural networks**

```python
import shap
import pandas as pd
import numpy as np
shap.initjs()
#Initialize JavaScript: Call shap.initjs() to enable the interactive visualizations.

customer = pd.read_csv("Customer Churn.csv")
customer.head()
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

X = customer.drop("Churn", axis=1) # Independent variables
y = customer.Churn # Dependent variable

# Split into train and test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Train a machine learning model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Make prediction on the testing data
y_pred = clf.predict(X_test)

# Classification Report
print(classification_report(y_pred, y_test))
```

```python
""" The model has shown better performance for ‚Äú0‚Äù label than ‚Äú1‚Äù due to an unbalanced dataset. 
Overall, it is an acceptable result with 94% accuracy."""
```

```python
"""Setting up SHAP Explainer Now comes the model explainer part.

We will first create an explainer object by providing a random forest classification model, then calculate SHAP value
using a testing set.

Create an Explainer: You initialize the SHAP Explainer with your trained model.

Calculate SHAP Values: You compute SHAP values for the test dataset, which allows you to analyze the impact of each feature 
on the predictions for the test data.

"""
```

```python
## Assuming clf is your trained model and X_test is your test data
# When you initialize the Explainer, it sets up the necessary computations and information needed to calculate 
# SHAP values for the model predictions.

explainer = shap.Explainer(clf)
shap_values = explainer.shap_values(X_test)
```

```python
# import image module 
from IPython.display import Image, display

display(Image(filename='shap.png', width=600, height=300))
```

```python
shap.summary_plot(shap_values, X_test)
#shap.summary_plot(shap_values, X_test, max_display=X_test.shape[1])
```

```python
"""  A positive SHAP value indicates that the feature increases the prediction (pushing it higher), 
while a negative SHAP value indicates that the feature decreases the prediction (pushing it lower)

."""
```

```python
shap.plots.bar(shap_values)
```

```python
shap.summary_plot(shap_values[0], X_test)
```

```python
shap.dependence_plot("Subscription Length", shap_values[0], X_test,interaction_index="Age")
```

```python
shap.decision_plot(explainer.expected_value[1], shap_values[1], X_test.columns)
```
